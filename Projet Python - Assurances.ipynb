```python
# 2. Nettoyage du dataset
```

```python
# 2.1 Suppression des doublons
```

```python
df_no_double = df.drop_duplicates()
print(df_no_double)
```

```python
# 2.2 Mise à jour du dataframe
```

```python
df_clean = df.copy()
```

```python
# 2.3 Remplacement des valeurs sur prime annuelle en 2 formules
```

```python
df_clean['prime_annuelle'] = df['prime_annuelle'].fillna (df['prime_annuelle'].median())
print(df_clean['prime_annuelle'])
```

```python
df_clean['prime_annuelle'] = df.groupby('type_contrat')['prime_annuelle'].transform(lambda x: x.fillna(x.median()))
print(df_clean['prime_annuelle'])
```

```python
# 2.4 Remplacement des valeurs sur scrore satisfaction en 2 formules
```

```python
df_clean['score_satisfaction'] = df_clean['score_satisfaction'].fillna(df_clean['score_satisfaction'].mean())
print(df_clean['score_satisfaction'])
```

```python
df_clean['score_satisfaction'] = df.groupby('conseiller')['score_satisfaction'].transform(lambda x: x.fillna(x.mean()))
print(df_clean['score_satisfaction'])
```

```python
# 2.5 Remplacement des valeurs sur conseiller
```

```python
df_clean['conseiller'] = df_clean['conseiller'].fillna('Non assigné')
print(df_clean['conseiller'])
```

```python
#2.6 Remplacement des valeurs sur age
```

```python
df_clean['age'] = df_clean['age'].fillna(df_clean['age'].median())
print(df_clean['age'])
```

```python
#2.7 Création de la colonne année de souscription
```

```python
import pandas as pd

df_clean['date_souscription'] = pd.to_datetime(df_clean['date_souscription'])
```

```python
df_clean['année_souscription'] = df_clean['date_souscription'].dt.year
print(df_clean['année_souscription'])
```

```python
#2.8 Création de la colonne mois de souscription
```

```python
df_clean['mois_souscription'] = df_clean['date_souscription'].dt.month_name()
print(df_clean['mois_souscription'])
```

```python
#2.9 Création de la colonne anciènneté en jour
```

```python
df_clean['ancienneté_jours'] = (pd.Timestamp.today() - df_clean['date_souscription']).dt.days
print(df_clean['ancienneté_jours'])
```

```python
#4 Performances des conseillers
```

```python
#4.1 Tableau recapitulatif par conseiller
```

```python
df_clean_perf_conseillers = df_clean.groupby('conseiller').agg ( Nb_contrats = ('type_contrat', 'count'),Prime_moy = ('prime_annuelle', 'mean') , Score_satisfaction = ('score_satisfaction', 'mean'),Nb_sinistres = ('nb_sinistres', 'sum'),
).round(2)
print(df_clean_perf_conseillers)
```

```python
#4.2 Taux de résiliation
```

```python
taux_resil = df_clean.groupby('conseiller')['statut'].apply( lambda x :round ((x == 'Résilié').sum() / len(x) * 100,1) ).rename('Taux_resiliation_%')
print(taux_resil)
```

```python
#4.3 Récap conseiller
```

```python
recap_conseiller = df_clean.groupby('conseiller').agg (prime_totale = ('prime_annuelle','sum'),chiffre_affaire = ('montant_sinistres','sum'),score_satisfaction = ('score_satisfaction','sum'),).round(2)
print(recap_conseiller)
```

```python
#4.4 Unir le tableau recapitulatif et le taux de résiliation
```

```python
df_clean_perf_conseillers = df_clean_perf_conseillers.join(taux_resil)
print(df_clean_perf_conseillers)
```

```python
#4.5 Trie par score de satisfaction décroissant
```

```python
df_clean_perf_conseillers = df_clean_perf_conseillers.sort_values('Score_satisfaction',ascending =False)
print(df_clean_perf_conseillers)
```
